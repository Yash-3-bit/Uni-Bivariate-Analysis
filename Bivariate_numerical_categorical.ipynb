```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.barplot(x='class',y='age',data=df)
```

```python
sns.barplot(x='class',y='age',data=df,estimator = 'sum')
```

```python
sns.barplot(x='class',y='fare',data=df,estimator = 'sum')
```

```python
sns.barplot(x='class',y='age',hue='sex',data=df)
```

```python
sns.barplot(x='class',y='fare',hue='sex',data=df)
```

```python
sns.boxplot(x='class',y='age',hue='sex',data=df)
```

```python
sns.distplot(df[df["class"] == "Second"]['age'])
```
